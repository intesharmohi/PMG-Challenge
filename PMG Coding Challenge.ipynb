```python
# Import packages
import pandas as pd


# Add filename column to the clothing data
df1 = pd.read_csv('clothing.csv')
df1["filename"] = 'clothing.csv'


# Add filename column to the accessories data
df2 = pd.read_csv('accessories.csv')
df2["filename"] = 'accessories.csv'


# Combine the two data frames into a single dataframe
combined_df = pd.merge(df1, df2, how='outer')


# Export the combined dataframe as a csv file
combined_df.to_csv('result.csv')
```
