```python
# Import packages
import os
import glob
import pandas as pd


# Set Directory
os.chdir(r"C:\Users\intes\Downloads\PMG")


# Add filename column to clothing.csv file
df1 = pd.read_csv('clothing.csv')
df1["filename"] = os.path.basename(r"C:\Users\intes\Downloads\PMG\clothing.csv")
df1.to_csv('clothing.csv', index=False)


# Add filename column to accessories.csv file
df2 = pd.read_csv('accessories.csv')
df2["filename"] = os.path.basename(r"C:\Users\intes\Downloads\PMG\accessories.csv")
df2.to_csv('accessories.csv', index=False)


# Match the csv file pattern using global
extension = '.csv'
all_files = [i for i in glob.glob(f"*{extension}")]


# Combine the files
combined_csv = pd.concat([pd.read_csv(f, encoding='utf-8-sig') for f in all_files])


# Export the combined file as a csv
combined_csv.to_csv('combined_csv.csv', index=False)
```
